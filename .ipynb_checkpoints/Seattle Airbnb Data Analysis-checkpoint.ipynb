## Which would be the best neighborhood to list an airbnb?

```python
seattle_shapes = gpd.read_file("./shape_files/City_Clerk_Neighborhoods.geojson")
seattle_shapes.head()
```

```python
seattle_shapes.plot()
```

```python
geo_listings = gpd.GeoDataFrame(listings, geometry=gpd.points_from_xy(listings.longitude, listings.latitude))
geo_listings.crs = seattle_shapes.crs
```

```python
ax = seattle_shapes.plot(color='white', edgecolor='black')
geo_listings.plot(ax=ax, color="green", markersize=1)
ax.set_axis_off()
```

```python
# using geopandas spatial join to determine each listing's neighborhood
geo_listings=gpd.sjoin(geo_listings, seattle_shapes[["S_HOOD", "geometry"]], op="within", how="left")
# calculating the number of aribnb for each neighborhood
seattle_shapes["number_of_airbnbs"] = seattle_shapes["S_HOOD"].map(geo_listings["S_HOOD"].value_counts().to_dict())
# calculating the average price for each neighborhood
seattle_shapes["price_of_airbnbs"] = seattle_shapes["S_HOOD"].map(geo_listings.groupby("S_HOOD").agg({"price":"mean"})["price"].to_dict())
```

```python
fig = plt.figure(figsize=(12,12), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.add_subplot(1,2,1)
ax.set_title("Heatmap of seattle by number of airbnbs")
seattle_shapes.plot(column="number_of_airbnbs", legend=True, ax=ax,  cmap="inferno")
ax = fig.add_subplot(1,2,2)
ax.set_title("Heatmap of seattle by avg price of airbnbs")
seattle_shapes.plot(column="price_of_airbnbs", legend=True, ax=ax,  cmap="inferno")
```

```python
kings_county_house_market= pd.read_csv("./shape_files/kingcounty/kc_house_data.csv")
# use the same method used over listings dataframe to transform lat and long columns to points and set that as the 
# geodataframe's geometry
kings_county_house_market = gpd.GeoDataFrame(kings_county_house_market, geometry=gpd.points_from_xy(kings_county_house_market.long, kings_county_house_market.lat))
kings_county_house_market.crs = seattle_shapes.crs
# asign a neighborhood to each house sale
kings_county_house_market = gpd.sjoin(kings_county_house_market, seattle_shapes[["S_HOOD", "geometry"]], op="within")

# Calculate average price per sqft of lots for each neighborhood
# dictionary with average price per sqft by neighborhood
f_calculate_price_persqft = lambda d: pd.Series({"avg_price_per_sqft": d["price"].sum()/d["sqft_lot"].sum()})
dict_of_avgpricepersqft = kings_county_house_market.groupby("S_HOOD").apply(f_calculate_price_persqft)["avg_price_per_sqft"].to_dict()
# asign to each neighborhood
seattle_shapes["avg_price_per_sqft"] = seattle_shapes["S_HOOD"].map(dict_of_avgpricepersqft)
# lets plot this data
fig = plt.figure(figsize=(12,12), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.add_subplot(1,2,1)
ax.set_title("House Sales")
ax.set_axis_off()
seattle_shapes.plot(color='white', edgecolor='black', ax=ax)
kings_county_house_market.plot(ax=ax, color="green", markersize=1)
ax = fig.add_subplot(1,2,2)
seattle_shapes.plot(column="avg_price_per_sqft", legend=True, cmap="inferno", ax=ax)
ax.set_title("Heatmap avg price per sqft")
ax.set_axis_off()
```

```python
# classifying each neighborhood in quantiles
seattle_shapes["rank_airbnb_density"] = pd.qcut(seattle_shapes["number_of_airbnbs"], 10, range(1,11))
seattle_shapes["rank_listing_price"] = pd.qcut(seattle_shapes["price_of_airbnbs"], 10, range(1,11))
# add rangom float between 0.001 and 0.009 to avoid the "bin edges must be unique error"
seattle_shapes["avg_price_per_sqft"] = seattle_shapes["avg_price_per_sqft"].apply(lambda p: p +random.uniform(0.001, 0.009))
seattle_shapes["rank_housing_price"] = pd.qcut(seattle_shapes["avg_price_per_sqft"], 10, range(1,11))
# Show the top 20
seattle_shapes.sort_values(["rank_airbnb_density", "rank_listing_price"], ascending=[False, False])[[
    "S_HOOD","rank_airbnb_density", "number_of_airbnbs", "rank_listing_price", "price_of_airbnbs", "rank_housing_price", "avg_price_per_sqft"]].head(20)
```

```python
# calculating top3 avg price per sqft
house_prices_2021_medians=[595000, 595000, 602500]# no data found for Central Business District,using the same as Belltown
house_prices_2016_medians=[485000, 485000, 320200]
avg_prices_per_sqft_2021 = [688, 792, 706]
top_3_indices= [59, 61, 62]
for i, idx in enumerate(top_3_indices):
    seattle_shapes.at[idx, "avg_price_per_sqft"] = avg_prices_per_sqft_2021[i] * house_prices_2016_medians[i] / house_prices_2021_medians[i]
# recalculate avg price per sqft quantiles
seattle_shapes["rank_housing_price"] = pd.qcut(seattle_shapes["avg_price_per_sqft"], 10, range(1,11))
# Show the top 20
seattle_shapes.sort_values(["rank_airbnb_density", "rank_listing_price"], ascending=[False, False])[[
    "S_HOOD","rank_airbnb_density", "number_of_airbnbs", "rank_listing_price", "price_of_airbnbs", "rank_housing_price", "avg_price_per_sqft"]].head(20).round(1)
```

```python
top_20_neighborhoods = seattle_shapes.sort_values(["rank_airbnb_density", "rank_listing_price"], ascending=[False, False])[[
    "S_HOOD","rank_airbnb_density", "number_of_airbnbs", "rank_listing_price", "price_of_airbnbs", "rank_housing_price", "avg_price_per_sqft"]].head(20).round(1)
```

## How much pricing varies per month?

```python
# month column
calendar["month"] = calendar["date"].dt.strftime('%b')
calendar["n_month"] =  calendar["date"].dt.month
# indicator for listings that have 12 months of data(at least one day on each month of the year)
months_available_by_listing = calendar[~calendar.price.isnull()].groupby("listing_id").agg({"n_month":lambda m: len(m.unique())})
calendar["has_12_months"] = calendar["listing_id"].map(months_available_by_listing.n_month.apply(lambda m: 1 if m ==12 else 0).to_dict())
# fill missing values for price with mean price for listing in month in order to be able to calculate z-score and percentage variation
monthly_mean_price_per_listing = calendar.groupby(["listing_id", "n_month"]).agg({"price":"mean"})
calendar["filled_price"] = calendar["price"]
for idx, row in calendar[calendar.price.isnull()].iterrows():
    try:
        calendar.at[idx, "filled_price"] = monthly_mean_price_per_listing.loc[(row["listing_id"], row["n_month"])]["price"]
    except:
        continue
# calculate z-score and percentage variation vs yearly median for each price
calendar["price_zunits"] = calendar.groupby("listing_id")["filled_price"].transform(lambda p: (p-p.mean())/p.std())
calendar["perc_relative_price_vs_median"] = calendar.groupby("listing_id")["filled_price"].transform(lambda p: (p-p.median())/p.median())
calendar.head()
```

```python
plt.figure(figsize=(15,6), dpi= 100)
ax = sns.boxplot(data=calendar.query("has_12_months==1").groupby(["n_month", "month", "listing_id"], as_index=False).agg({"price_zunits":"median"}), x="month", y="price_zunits", showfliers=False, color="lightsteelblue")
ax.set(xticks=range(12), xticklabels=calendar.month.unique())
ax.set(title="Listing Monthly Median Price Zscore Distribution", xlabel="Months")
```

```python
fig = plt.figure(figsize=(15,6), dpi= 100)
ax = fig.add_subplot()
ax = sns.lineplot(data=calendar.query("has_12_months==1"), x="n_month", y="perc_relative_price_vs_median", label="Percentage Variation vs Yearly Median Price")
ax.set(ylabel="Percentage Variation vs Yearly Median Price")
ax.legend(loc="upper left")
ax.yaxis.set_major_formatter(PercentFormatter(1))
ax2 = ax.twinx()
# Monthly Temperatures from https://www.tripsavvy.com/the-weather-and-climate-in-seattle-4175384
sns.lineplot(y=[48,48,54,59,66,71,78,79,71,60,52,46], x=range(1,13), color="red", label="Temperature", legend=False)
ax.set(xticks=calendar.n_month.unique(), xticklabels=list(calendar.month.unique()), xlabel="Months")
ax2.set_ylabel("Temperature F°")
ax2.legend(loc="upper right")
plt.title("Mean Price Percentage Variation Compared to Temperature")
```

## What are the key factors for an airbnb listing to be successful?

```python
#filter only listings that have 300 days or more availble per year to be able to compare them
listings_success= geo_listings.query("availability_365 >300")
# defining success as the multiplication of the number of reviews times the average overall review score
listings_success["success"] = listings_success["number_of_reviews"] * listings_success["review_scores_rating"]
# plot the distributions to see if there's a difference
plt.figure(figsize=(8,8), dpi= 100)
sns.histplot(data=listings_success, x="success", hue="host_is_superhost", stat="density", common_norm=False) 
```

```python
# Perform the Mann-Whitney U rank test
u, p_value = mannwhitneyu(listings_success.query("host_is_superhost =='f'")["success"], 
                          listings_success.query("host_is_superhost =='t'")["success"], 
                          alternative="less")
print(p_value)
```
